```python
def g(x):
    return x ** 3 + 1

def gprime(x):
    return 3 * x**2

def solve_x3(x0, eps=1e-2):
    n_iter = 0
    x_curr = x0
    while n_iter < 20:
        n_iter += 1
        x_next = x_curr
        x_curr = x_curr-g(x_curr)/gprime(x_curr)
        if (abs(x_curr-x_next)) < eps:
            break
    return x_curr
solve_x3 (-6+6j)
```
